```python
# Part 2 

deck=pd.read_csv("./data/input.txt", sep=" ", header=None)
deck.columns = ['hand', 'bid']

def score_hand_with_jokers(hand):
    
    hand_no_jokers = hand.replace("J", "")
    num_jokes = len(hand) - len(hand_no_jokers)
    if num_jokes == 5: # cant add jokers to an empty counter.
        return 6

    count = Counter(list(hand_no_jokers))
    count[count.most_common(1)[0][0]] += num_jokes
    num_mc = count.most_common(1)[0][1]
    unique_cards = len(count)
    if unique_cards == 1:
        return 6# five of a kind
    elif unique_cards == 2 and num_mc == 4:
        return 5# four of a kind
    elif unique_cards == 2 and num_mc == 3:
        return 4# full house
    elif num_mc == 3:
        return 3# 3 of a kind
    elif num_mc == 2 and unique_cards ==3:
        return 2 # 2 pair
    elif num_mc == 2:
        return 1# a pair 
    else:
        return 0# highcard


deck['score'] = deck.hand.apply(score_hand_with_jokers)

# jacks are now jokers and arent worth as much, so need a new hex map
hex_map = {
    "A": "e", "K":"d", "Q":"c", "J": "1", "T": "a"
}

def convert_to_hex(hand):
    hand = list(hand)
    hex_out = []
    for card in hand:
        if card in hex_map:
            hex_out.append(hex_map[card])
        else:
            hex_out.append(card)

    return hex(int("".join(hex_out), 16))


deck['hex'] = deck.hand.apply(convert_to_hex)
deck


deck=deck.sort_values(['score', 'hex'], ascending=True).reset_index()
deck['rank'] = deck.index+1

sum(deck['rank']*deck['bid'])
```
