```python
with open("./data/input.txt") as f:
    coord = f.readlines()

instructions = list(coord.pop(0).strip())
instructions

# {for step in coord if step != "\n"}
map = {}

for row in coord:
    if row != "\n":
        step = row.replace(r"(", "").replace(r")", "").replace(r"=", "").replace(",", "").split()
        
        map.update(
            {step[0]: {"L": step[1], "R": step[2]}}
        )

map
```

```python
def count_steps(instructions, map):
    not_found = True
    start = "AAA"
    num_steps = 0
    while not_found:
        for instruct in instructions:
            num_steps += 1
            start = map[start][instruct]

            if start == 'ZZZ':
                return num_steps

print(count_steps(instructions, map))
    
    
```
